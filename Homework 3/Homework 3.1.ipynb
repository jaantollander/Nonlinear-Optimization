## (a)

Function $f$ is defined
$$
f(x) = x_1^3 - x_1 + x_2^3 - x_2.
$$

```julia
f(x_1, x_2) = x_1^3 - x_1 + x_2^3 - x_2;
```

```julia
x_1 = -3:0.1:3
x_2 = -3:0.1:3;
```

Surface and contour plots of function $f$.

```julia
plot(x_1, x_2, f, st=:surface)
```

```julia
plot(x_1, x_2, f, st=:contour, levels = 200)
```

The function $f$ is not convex.

## (b)

The first-order necessary condition, that is, the gradient equal zero $\nabla f(x) = 0$ is

$$
\displaystyle \left[\begin{matrix}3 x_{1}^{2} - 1\\3 x_{2}^{2} - 1\end{matrix}\right] = \left[\begin{matrix}0\\0\end{matrix}\right].
$$

By solving is we receive the following four critical points
$$
\displaystyle \left[ \left\{ x_{1} : - \frac{\sqrt{3}}{3}, \  x_{2} : - \frac{\sqrt{3}}{3}\right\}, \  \left\{ x_{1} : - \frac{\sqrt{3}}{3}, \  x_{2} : \frac{\sqrt{3}}{3}\right\}, \  \left\{ x_{1} : \frac{\sqrt{3}}{3}, \  x_{2} : - \frac{\sqrt{3}}{3}\right\}, \  \left\{ x_{1} : \frac{\sqrt{3}}{3}, \  x_{2} : \frac{\sqrt{3}}{3}\right\}\right].
$$

## (c)

We can also calculate the Heassian $H(x)$ of $f$ to be

$$
\displaystyle \left[\begin{matrix}6 x_{1} & 0\\0 & 6 x_{2}\end{matrix}\right]
$$

By subsituting the values of the critical points, we receive

$$
\displaystyle \left[ \left[\begin{matrix}- 2 \sqrt{3} & 0\\0 & - 2 \sqrt{3}\end{matrix}\right], \  \left[\begin{matrix}- 2 \sqrt{3} & 0\\0 & 2 \sqrt{3}\end{matrix}\right], \  \left[\begin{matrix}2 \sqrt{3} & 0\\0 & - 2 \sqrt{3}\end{matrix}\right], \  \left[\begin{matrix}2 \sqrt{3} & 0\\0 & 2 \sqrt{3}\end{matrix}\right]\right]
$$

The eigenvalues and vectors of each of the Hessian matrices at the critical points

```julia
eigvals([-2*√3 0; 0 -2*√3])
```

```julia
eigvals([-2*√3 0; 0 2*√3])
```

```julia
eigvals([2*√3 0; 0 -2*√3])
```

```julia
eigvals([2*√3 0; 0 2*√3])
```

The Hessian matrix of a convex function is positive semi-definite. As can be seen from the eigenvalues, the Hessian is indefinite, thus $f$ is not convex. 

From the eigenvalues of the critical points, we can also see that the function $f$ has one local minimum (negative eigenvalues), one local maximum (positive eigenvalues) and two saddle points (one positive, one negative eigenvalue).
